```python
# Compute means over assets and sectors dimensions for each point in the earnings grid
means_over_z_grid = np.mean(model.ss.c, axis=(0, 2))  # taking mean over the first and third dimensions

# Plot
plt.figure(figsize=(10, 6))
for i in range(means_over_z_grid.shape[0]):  # iterating over the earnings grid dimension
    plt.plot(means_over_z_grid[i, :], label=f'z_grid={model.par.z_grid[i]}')

plt.title("Mean consumption over assets and sectors for each point in the earnings grid")
plt.xlabel("beta")
plt.ylabel("Mean value")
plt.legend()
plt.show()
```
